```python
#importing libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading the dataset
df = pd.read_csv("Arabica datasetssss.csv")
df.head()
```

```python
# Shape and basic info
print("Shape:", df.shape)
df.info()
df.describe(include='all')
```

```python
# Check nulls
print(df.isnull().sum())

# Fill missing numeric values with median
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Fill missing text values with mode
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Encode all categorical values to numbers
for col in cat_cols:
    df[col] = df[col].astype('category').cat.codes
```

```python
#Normalize Numerical Columns (0 to 1 scale)
for col in num_cols:
    df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
```

```python
# Simple boxplot for numeric columns  Visualize Outliers
for col in num_cols:
    plt.figure(figsize=(6, 1.5))
    plt.boxplot(df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
#Remove Outliers
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
#Save Cleaned Dataset
df.to_csv("cleaned_arabica_dataset.csv", index=False)
print("Saved cleaned data as cleaned_arabica_dataset.csv")
```
